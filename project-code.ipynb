```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('Dataset/all_kendrick_tracks.csv')
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.sort_values(by='release_date', ascending= False)
```

```python
df
```

```python
df['loudness'] = df['loudness'] * -1
```

```python
df
```

```python
df_yearly = df.groupby('release_year').agg({
    'track_id': 'count',          # number of tracks released that year
    'popularity': 'mean',         # average popularity
    'danceability': 'mean',
    'energy': 'mean',
    'valence': 'mean',
    'tempo': 'mean',
    'loudness': 'mean',
    'duration_min': 'mean',       # average duration in minutes
})

```

```python
df_yearly = df_yearly.rename(columns={'track_id': 'track_count'})
```

```python
df_yearly
```

```python
df_album = df.groupby('album_name').agg({
    'track_id': 'count',          
    'popularity': 'mean',         
    'danceability': 'mean',
    'energy': 'mean',
    'valence': 'mean',
    'tempo': 'mean',
    'loudness': 'mean',
    'duration_min': 'mean',
    'release_year': 'first'      
})


```

```python
df_album = df_album.rename(columns={'track_id': 'track_count'})
```

```python
df_album
```

```python
df_album = df_album.sort_values(by='release_year', ascending= False)
```

```python
df_album
```

```python
df_album_count = df_album.sort_values('track_count', ascending=False).head(20)
```

```python
plt.figure(figsize=(10, 10))
plt.bar(df_album_count.index, df_album_count['track_count'])
plt.xticks(rotation=90)
plt.ylabel("Number of Tracks")
plt.title("Highest Tracks Per Album")
plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_yearly.index, df_yearly['popularity'], marker='o')
plt.xlabel("Release Year")
plt.ylabel("Average Popularity")
plt.title("Average Popularity Over the Years")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_yearly.index, df_yearly['track_count'], marker='o')
plt.xlabel("Release Year")
plt.ylabel("Track Count")
plt.title("Tracks Released Over the Years")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.plot(df_yearly.index, df_yearly['energy'], label='Tempo', marker='o')
plt.plot(df_yearly.index, df_yearly['danceability'], label='Tempo', marker='o')
plt.plot(df_yearly.index, df_yearly['valence'], label='Valence', marker='o')

plt.xlabel("Release Year")
plt.ylabel("Value")
plt.title("Audio Attributes Over Years")
plt.legend()  
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
X = df_yearly.index.values.reshape(-1, 1)  
y = df_yearly['popularity'].values          
```

```python
lin_model = LinearRegression()
lin_model.fit(X, y)
```

```python
# Cell 4: Predict future popularity for next 3 years
last_year = df_yearly.index.max()
future_years = np.array([last_year + 1, last_year + 2, last_year + 3]).reshape(-1, 1)

future_predictions = lin_model.predict(future_years)

future_df = pd.DataFrame({
    "year": future_years.flatten(),
    "predicted_popularity": future_predictions
})

future_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

plt.plot(df_yearly.index, df_yearly['popularity'], marker='o', label='Historical Popularity')

x_all = np.arange(df_yearly.index.min(), last_year + 4).reshape(-1, 1)
y_all_pred = lin_model.predict(x_all)

plt.scatter(future_df['year'], future_df['predicted_popularity'],
            color='orange', label="Future Predictions")

plt.title("Popularity Forecast — Linear Regression")
plt.xlabel("Year")
plt.ylabel("Popularity")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))

x_historical = df_yearly.index.values
x_future = future_df['year'].values
x_combined = np.concatenate([x_historical, x_future])

y_historical = df_yearly['popularity'].values
y_future = future_df['predicted_popularity'].values
y_combined = np.concatenate([y_historical, y_future])

plt.plot(x_combined, y_combined, marker='o', label='Historical Actuals & Future Forecast (Combined Line)')


plt.title("Final Popularity Forecast")
plt.xlabel("Year")
plt.ylabel("Popularity")
plt.legend()
plt.grid(True)
plt.show()
```
