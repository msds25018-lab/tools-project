```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
X = df_yearly.index.values.reshape(-1, 1)  
y = df_yearly['popularity'].values          
```

```python
lin_model = LinearRegression()
lin_model.fit(X, y)
```

```python
last_year = df_yearly.index.max()
future_years = np.array([last_year + 1, last_year + 2, last_year + 3]).reshape(-1, 1)

future_predictions = lin_model.predict(future_years)

future_df = pd.DataFrame({
    "year": future_years.flatten(),
    "predicted_popularity": future_predictions
})

future_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
import matplotlib.pyplot as plt

df_ts = df_yearly.copy()
```

```python

df_ts['popularity_lag1'] = df_ts['popularity'].shift(1)
df_ts['popularity_lag2'] = df_ts['popularity'].shift(2)

df_ts['energy_lag1'] = df_ts['energy'].shift(1)
df_ts['danceability_lag1'] = df_ts['danceability'].shift(1)

df_ts['popularity_roll3'] = df_ts['popularity'].rolling(3).mean()

df_ts = df_ts.dropna()
```

```python

feature_cols = [
    'track_count',
    'danceability', 'energy', 'valence', 'tempo', 'loudness',
    'duration_min',
    'popularity_lag1', 'popularity_lag2',
    'energy_lag1', 'danceability_lag1',
    'popularity_roll3',
    'year'  
]

df_ts['year'] = df_ts.index

X = df_ts[feature_cols]
y = df_ts['popularity']
X
y
```

```python
model = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='reg:squarederror'
)

model.fit(X, y)
```

```python
future_years = [df_ts.index.max() + 1,
                df_ts.index.max() + 2,
                df_ts.index.max() + 3]

future_preds = []

last_row = df_ts.iloc[-1].copy()

for year in future_years:

    row = pd.DataFrame([{
        'track_count': last_row['track_count'],
        'danceability': last_row['danceability'],
        'energy': last_row['energy'],
        'valence': last_row['valence'],
        'tempo': last_row['tempo'],
        'loudness': last_row['loudness'],
        'duration_min': last_row['duration_min'],
        'popularity_lag1': last_row['popularity_lag1'],
        'popularity_lag2': last_row['popularity_lag2'],
        'energy_lag1': last_row['energy_lag1'],
        'danceability_lag1': last_row['danceability_lag1'],
        'popularity_roll3': last_row['popularity_roll3'],
        'year': year
    }])

    pred = model.predict(row)[0]
    future_preds.append(pred)

    last_row['popularity_lag2'] = last_row['popularity_lag1']
    last_row['popularity_lag1'] = pred

    last_row['popularity_roll3'] = np.mean([
        last_row['popularity_lag1'],
        last_row['popularity_lag2'],
        df_ts['popularity'].iloc[-1]  
    ])
```

```python
future_df = pd.DataFrame({
    "year": future_years,
    "predicted_popularity": future_preds
})

print(future_df)

```

```python
historical_fit = model.predict(X)

plt.figure(figsize=(12, 6))
```

```python


plt.plot(
    df_yearly.index,
    df_yearly['popularity'],
    marker='o',
    linewidth=2,
    label='Historical Actual'
)

plt.plot(
    df_ts.index,                  
    historical_fit,               
    marker='o',
    linestyle='--',
    linewidth=2,
    label='Model Fit (XGBoost)'
)

plt.plot(
    future_df['year'],
    future_df['predicted_popularity'],
    marker='o',
    linewidth=2,
    label='Forecast (Next 3 Years)'
)

plt.title("Kendrick Popularity Forecast (XGBoost)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Popularity", fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
