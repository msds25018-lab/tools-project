```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
import matplotlib.pyplot as plt

df_ts = df_yearly.copy()

# ------------------------------------------------
# CREATE LAGS + ROLLING FEATURES
# ------------------------------------------------
df_ts['popularity_lag1'] = df_ts['popularity'].shift(1)
df_ts['popularity_lag2'] = df_ts['popularity'].shift(2)

df_ts['energy_lag1'] = df_ts['energy'].shift(1)
df_ts['danceability_lag1'] = df_ts['danceability'].shift(1)

df_ts['popularity_roll3'] = df_ts['popularity'].rolling(3).mean()

# Drop NA rows
df_ts = df_ts.dropna()

# ------------------------------------------------
# SELECT EXACT FEATURES FOR XGBOOST
# ------------------------------------------------
feature_cols = [
    'track_count',
    'danceability', 'energy', 'valence', 'tempo', 'loudness',
    'duration_min',
    'popularity_lag1', 'popularity_lag2',
    'energy_lag1', 'danceability_lag1',
    'popularity_roll3',
    'year'  # add year at the end
]

# Add year column
df_ts['year'] = df_ts.index

X = df_ts[feature_cols]
y = df_ts['popularity']

# ------------------------------------------------
# TRAIN MODEL
# ------------------------------------------------
model = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='reg:squarederror'
)

model.fit(X, y)

# ------------------------------------------------
# FORECAST NEXT 3 YEARS USING AUTO-REGRESSIVE LOOP
# ------------------------------------------------
future_years = [df_ts.index.max() + 1,
                df_ts.index.max() + 2,
                df_ts.index.max() + 3]

future_preds = []

last_row = df_ts.iloc[-1].copy()

for year in future_years:

    # Build row with EXACT SAME COLUMNS
    row = pd.DataFrame([{
        'track_count': last_row['track_count'],
        'danceability': last_row['danceability'],
        'energy': last_row['energy'],
        'valence': last_row['valence'],
        'tempo': last_row['tempo'],
        'loudness': last_row['loudness'],
        'duration_min': last_row['duration_min'],
        'popularity_lag1': last_row['popularity_lag1'],
        'popularity_lag2': last_row['popularity_lag2'],
        'energy_lag1': last_row['energy_lag1'],
        'danceability_lag1': last_row['danceability_lag1'],
        'popularity_roll3': last_row['popularity_roll3'],
        'year': year
    }])

    pred = model.predict(row)[0]
    future_preds.append(pred)

    # --- UPDATE LAGS FOR NEXT ITERATION ---
    last_row['popularity_lag2'] = last_row['popularity_lag1']
    last_row['popularity_lag1'] = pred

    last_row['popularity_roll3'] = np.mean([
        last_row['popularity_lag1'],
        last_row['popularity_lag2'],
        df_ts['popularity'].iloc[-1]  # safe fallback
    ])

# Create future dataframe
future_df = pd.DataFrame({
    "year": future_years,
    "predicted_popularity": future_preds
})

print(future_df)

# ------------------------------------------------
# PLOT RESULT
# ------------------------------------------------
plt.figure(figsize=(10,5))

plt.plot(df_yearly.index, df_yearly['popularity'], marker='o', label='Historical')
plt.plot(X['year'], model.predict(X), label='XGBoost Fit')

plt.scatter(future_df['year'], future_df['predicted_popularity'],
            color='orange', marker='o', label='Future Forecast')

plt.title("Kendrick Popularity Forecast (XGBoost with Lag Features)")
plt.xlabel("Year")
plt.ylabel("Popularity")
plt.grid(True)
plt.legend()
plt.show()
```

ML -Part ARIMA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pmdarima import auto_arima
from statsmodels.tsa.arima.model import ARIMA

# -------------------------
# 1. Prepare the Data
# -------------------------

# reset index so 'release_year' becomes a column
df_arima = df_yearly.reset_index()

# keep only what we need
df_arima = df_arima[['release_year', 'popularity']]

# rename and set index
df_arima = df_arima.rename(columns={'release_year': 'year'})
df_arima = df_arima.set_index('year')

# convert index to datetime
df_arima.index = pd.to_datetime(df_arima.index.astype(int), format='%Y')

# -------------------------
# 2. Fit ARIMA Model
# -------------------------

model_auto = auto_arima(
    df_arima['popularity'],
    seasonal=False,
    trace=True,
    suppress_warnings=True,
    stepwise=True
)

print("\nBest ARIMA order:", model_auto.order)

model = ARIMA(df_arima['popularity'], order=model_auto.order)
model_fit = model.fit()

print(model_fit.summary())

# -------------------------
# 3. Forecast next 3 years
# -------------------------

n_years = 3
forecast = model_fit.forecast(steps=n_years)

future_years = pd.date_range(
    start=df_arima.index.max() + pd.offsets.YearBegin(),
    periods=n_years,
    freq='YS'
)

forecast_df = pd.DataFrame({
    'year': future_years.year,
    'predicted_popularity': forecast.values
})

print("\nForecast:")
print(forecast_df)

# -------------------------
# 4. Plot
# -------------------------

plt.figure(figsize=(10,5))
plt.plot(df_arima.index.year, df_arima['popularity'], label='Historical')
plt.plot(forecast_df['year'], forecast_df['predicted_popularity'],
         marker='o', linestyle='--', label='Forecast')
plt.xlabel("Year")
plt.ylabel("Popularity")
plt.title("ARIMA Forecast of Popularity")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Install (uncomment if needed)
# !pip install statsmodels

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings("ignore")

# -----------------------
# Prepare the series
# -----------------------
# Expect df_yearly to exist and have index = years and a 'popularity' column
# If your df_yearly is a DataFrame with index as year (int) and column 'popularity', we're good.
# Example:
# df_yearly.head()

series = df_yearly['popularity'].copy()
# convert index to datetime for plotting + model compatibility
series.index = pd.to_datetime(series.index.astype(int).astype(str), format='%Y')

# -----------------------
# Optional: small model selection by AIC (safe for very short series)
# -----------------------
# Try a few combinations and pick the best AIC. This is optional — skip or use only one model if preferred.
candidates = []
for trend in [None, 'add', 'mul']:
    for damped in [False, True]:
        if trend is None and damped:
            # can't use damped when there's no trend
            continue
        try:
            m = ExponentialSmoothing(series, trend=trend, damped_trend=damped,
                                     seasonal=None, initialization_method='estimated')
            fit = m.fit(optimized=True)
            candidates.append((trend, damped, fit.aic, fit))
        except Exception as e:
            # ignore combinations that fail (very short series can cause failures)
            pass

if len(candidates) == 0:
    raise RuntimeError("No ETS candidate could be fit on your data. Consider using a simpler model or more observations.")

# pick best by AIC
candidates = sorted(candidates, key=lambda x: x[2])
best_trend, best_damped, best_aic, best_fit = candidates[0]
print(f"Selected ETS model: trend={best_trend}, damped={best_damped}, AIC={best_aic:.2f}")

# Use the selected fit object
fit = best_fit

# -----------------------
# Optional: Holdout evaluation
# -----------------------
# If you want a quick backtest, hold out last k years (here 3)
k = 3
if len(series) > k + 1:
    train = series.iloc[:-k]
    test = series.iloc[-k:]
    # fit on train
    m2 = ExponentialSmoothing(train, trend=best_trend, damped_trend=best_damped,
                              seasonal=None, initialization_method='estimated')
    fit2 = m2.fit(optimized=True)
    pred_test = fit2.forecast(steps=k)
    rmse = np.sqrt(mean_squared_error(test, pred_test))
    print(f"HOLDOUT RMSE (last {k} years): {rmse:.3f}")
else:
    print("Skipping holdout evaluation because series is too short.")

# -----------------------
# Forecast next N years
# -----------------------
n_forecast = 3
forecast = fit.forecast(steps=n_forecast)  # returns a pandas Series with datetime index

# Build a simple approximate 95% prediction interval using residual std (approximate)
resid = fit.fittedvalues - series  # in-sample residuals (aligned by index)
sigma = resid.std(ddof=0)  # population std
z = 1.96  # ~95% CI
upper = forecast + z * sigma
lower = forecast - z * sigma

# Prepare printable dataframe for forecast (years as integers)
forecast_df = pd.DataFrame({
    'year': forecast.index.year,
    'predicted_popularity': forecast.values,
    'lower_95': lower.values,
    'upper_95': upper.values
})
print("\nForecast (approx. 95% interval):")
print(forecast_df)

# -----------------------
# Plot historical + forecast + interval
# -----------------------
plt.figure(figsize=(10,5))
plt.plot(series.index.year, series.values, marker='o', label='Historical (popularity)')
# plot fitted values (in-sample fit)
plt.plot(fit.fittedvalues.index.year, fit.fittedvalues.values, label='Fitted (in-sample)', alpha=0.7)

# forecast line + points
plt.plot(forecast.index.year, forecast.values, marker='o', linestyle='--', label='Forecast')

# shaded interval
plt.fill_between(forecast.index.year, lower.values, upper.values, color='gray', alpha=0.25, label='Approx 95% CI')

plt.xlabel('Year')
plt.ylabel('Popularity')
plt.title('ETS Forecast (Exponential Smoothing) of Popularity')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
